```python
import re
from bs4 import BeautifulSoup
import requests
import csv
import sys

# Set the encoding to UTF-8 to handle Unicode characters
if sys.version_info < (3, 8):
    sys.stdout = open(sys.stdout.fileno(), mode='w', encoding='utf-8', buffering=1)

# Number of times to run the code
num_runs = 562# as trustpilot atombank has 562 pages

# Create a CSV file and write the header row with UTF-8 encoding (outside the loop)
with open("E:/Trustpilot_reviews/Customer_reviews.csv", "w", newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(["Name", "Rating", "Review Title", "Review Text"])

    for run in range(0, num_runs + 1):  # Start the range from 1
        if run ==0:
            url = f"https://www.trustpilot.com/review/atombank.co.uk"
        else:
            url = f"https://www.trustpilot.com/review/atombank.co.uk?page={run}"
            # Connect to the website and get the HTML content
            scrape = requests.get(url).text

            # Parse the HTML content using BeautifulSoup
            soup = BeautifulSoup(scrape, 'html.parser')

            # Find all the article tags with the specified class
            maintags = soup.find_all('article', class_="paper_paper__1PY90 paper_outline__lwsUX card_card__lQWDv card_noPadding__D8PcU styles_reviewCard__hcAvl")

            # Loop through the main tags and extract data
            for i in maintags:
                # Extract the desired information
                name = i.find('span', {"data-consumer-name-typography": "true"}).text
                rating_string = i.find('div', class_="star-rating_starRating__4rrcf star-rating_medium__iN6Ty").img.get("alt")
                review_title = i.find('h2', class_="typography_heading-s__f7029 typography_appearance-default__AAY17").text

                # Use regular expression to extract the numeric rating
                rating = re.search(r'\d+', rating_string)

                # Check if a rating was found and extract it as an integer
                if rating:
                    rating_value = int(rating.group())
                else:
                    rating_value = None  # Handle cases where no rating is found

                # Find the review text and handle cases where it's not found
                review_text_elem = i.find('p', {"data-service-review-text-typography": "true"})
                if review_text_elem:
                    review_text = review_text_elem.text
                else:
                    review_text = ""

                # Write the data to the CSV file
                csv_writer.writerow([name.strip(), rating_value, review_title.strip(), review_text.strip()])

            print(f"Run {run} completed.")
```
